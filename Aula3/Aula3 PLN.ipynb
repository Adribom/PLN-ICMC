# Exercício1 da Aula 3 de PLN

1. Dado o arquivo qbdata.txt, retorne o rating de
cada QB na forma “nome do QB teve valor XX.X’
e escreva em um arquivo novo.


```python
# Abrir os arquivos de entrada e de saída
infile = open("/content/drive/MyDrive/Python para Processamento de Linguagem Natural/Aula3/qbdata.txt",'r')
outfile = open("/content/drive/MyDrive/Python para Processamento de Linguagem Natural/Aula3/resultsEx1.txt", 'w')
```

```python
content = infile.read()
print(f"Conteúdo do infile:\n\n{content}")
print(f"\nTipo da variável content: {type(content)}")
```

```python
# Separar os dados da string por linha
split_content = content.split('\n')

# Modificar cada string da lista para manter apenas os dados necessários
# x.replace(x[x.find("QB"):x.find("%")], '')
final_content = [f'{x.replace(x[x.find("QB"):x.find("%") + 1], "teve valor")}' for x in split_content]
```

```python
# Juntar novamente a uma só string
final_string = "\n".join(final_content)

# Escrever no arquivo de saída
outfile.write(final_string)


infile.close()
outfile.close()
```
