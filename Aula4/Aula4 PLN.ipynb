## Tokenização
Separar as palavras do texto (parecido com um split())

Palavra, número, sinal de pontuação = Token

```python
nltk.word_tokenize("Com um passe de Eli Manning para Plaxico Burress a 39 segundos do fim, o New York Giants anotou o touchdown decisivo e derrubou o favorito New England Patriots por 17 a 14 neste domingo, em Glendale, no Super Bowl XLII.")
```

### Tokenizar sem sinais de pontuação

```python
from nltk.tokenize import RegexpTokenizer
texto = "Com um passe de Eli Manning para Plaxico Burress a 39 segundos do fim, o New York Giants anotou o touchdown decisivo e derrubou o favorito New England Patriots por 17 a 14 neste domingo, em Glendale, no Super Bowl XLII."
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(texto)
print(tokens)
```

### Tokenizar sem números

```python
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w*')
tokens_palavras = tokenizer.tokenize(texto)
print(tokens_palavras)
```

## Frequência/Contagem de palavras
Ordena a Frequência dos tokens


```python
texto = "Com um passe de Eli Manning para Plaxico Burress a 39 segundos do fim, o New York Giants anotou o touchdown decisivo e derrubou o favorito New England Patriots por 17 a 14 neste domingo, em Glendale, no Super Bowl XLII."

tokenizer  = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(texto)
frequencia = nltk.FreqDist(tokens)

print(frequencia.most_common(10)) # Mostrar a palavras com maior frequencia, neste caso, as top 10
```

## *Stopwords*
Palavras que não têm valor para a aplicação procurada

```python
stopwords = nltk.corpus.stopwords.words('portuguese') # Mostras todas as palavras consideradas como stopwords na língua portuguesa
```

```python
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w*')
tokens = tokenizer.tokenize(corpus)

nova_lista = []

# Colocar todas as palavras em caixa baixa e remover as stopwords
for token in tokens:
  if token.lower() not in stopwords:
    nova_lista.append(token.lower())

frequencia = nltk.FreqDist(nova_lista)
frequencia.most_common(10)
```

```python
# Com List Comprehension

tokenizer = RegexpTokenizer(r'[a-zA-Z]\w*')
tokens = tokenizer.tokenize(corpus)

# Colocar todas as palavras em caixa baixa e remover as stopwords
nova_lista = [token.lower() for token in tokens if token.lower() not in stopwords]

frequencia = nltk.FreqDist(nova_lista)
frequencia.most_common(10)
```

## Stemmer e Lemmatizer

### ***Stemming***: 
reduzir a palavra ao ser **radical**
- **amig**: amigo, amiga, amigão 

### ***Lematização***: 
reduzir a palavra à sua **forma canônica**, levando em conta sua **classe gramatical**
- **estudar**: estudando, estudioso, estudei

**OBS:** A Lematização no NLTK é boa para o inglês e ruim para o Português 

#### RSLP - Removedor de Sufixos da Língua Potuguesa

```python
stemmer = nltk.RSLPStemmer()

print(stemmer.stem("Amigão"))
print(stemmer.stem("amigo"))
print(stemmer.stem("amigos"))
```

## Etiquetadores
Dá a classe gramatical de uma palavra.

### Etiquetador pelo corpus MacMorpho

```python
from nltk.corpus import mac_morpho
from nltk.tag import UnigramTagger

tokens = nltk.word_tokenize(corpus)

sentencas_treino = mac_morpho.tagged_sents()
etiquetador = UnigramTagger(sentencas_treino)

etiquetado = etiquetador.tag(tokens)

print(etiquetado)
```

```python
#Usando uma etiqueta de escape para quando aparecer uma palavra que não é reconhecida por não estar no dataset
from nltk.corpus import mac_morpho
from nltk.tag import UnigramTagger
from nltk.tag import DefaultTagger

tokens = nltk.word_tokenize(corpus)

etiq_padrao = DefaultTagger('N')
sentencas_treino = mac_morpho.tagged_sents()
etiquetador = UnigramTagger(sentencas_treino, backoff=etiq_padrao)

etiquetado = etiquetador.tag(tokens)

print(etiquetado)
```

```python
# Agrupar quando houver classificações iguais seguidas
from nltk.chunk import RegexpParser

pattern = 'NP: {<NPROP><NPROP> | <N><N>}'
analise_gramatical = RegexpParser(pattern)

arvore = analise_gramatical.parse(etiquetado)
print(arvore)
```
