## Carregar o texto e deixá-lo analisado pelo spacy em uma variável

```python
import spacy

texto = "A Comissão Parlamentar de Inquérito (CPI) da Covid ouve nesta quarta-feira (29) o empresário bolsonarista Luciano Hang, dono da rede de lojas Havan, suspeito de financiar a divulgação de notícias falsas, em especial sobre tratamentos ineficazes contra a Covid-19."
nlp = spacy.load("pt_core_news_lg")
doc = nlp(texto)
```

```python
tokens = [token for token in doc]
tokens
```

```python
# Retornar os objetos transformados em strings
tokens = [token.orth_ for token in doc]
tokens
```

```python
# Retornar apenas os alphas(palavras) de um texto
alpha_tokens = [token.orth_ for token in doc if token.is_alpha]
print(f'Alpha Tokens: {alpha_tokens}')


# Retornar apenas os digitos de um texto
digit_tokens = [token.orth_ for token in doc if token.is_digit]
print(f'Digit Tokens: {digit_tokens}')

# Retornar apenas as pontuações de um texto
punct_tokens = [token.orth_ for token in doc if token.is_punct]
print(f'Punct Tokens: {punct_tokens}')
```

```python
import spacy

nlp = spacy.load("pt_core_news_lg")
doc = nlp(corpus)

tokens = [token.orth_ for token in doc]
```

## Lematização

```python
lemmas = [token.lemma_ for token in doc if token.pos_ == 'VERB'] ## Trazer os verbos
```

```python
# Lista com os tokens e suas respectivas classes
pos =  [(token.orth_, token.pos_) for token in doc]
pos
```

## Etiquetador

```python
morfologicas = [(token.orth_, token.morph) for token in doc]
morfologicas
```

## Entidades nomeadas

```python
entidades_nomeadas = list(doc.ents)
print(entidades_nomeadas)

detalhes_entidades = [(entidade, entidade.label_) for entidade in doc.ents]
detalhes_entidades
```

## DisplaCy

```python
# Ecrever em um html, fica bonitinho
html = spacy.displacy.render(doc, style="ent")
output_path = open('entidades_nomeadas.html', 'w', encoding='utf-8')
output_path.write(html)
output_path.close()
```

```python
sintaxe = [(token.orth_, token.dep_) for token in doc]
```
